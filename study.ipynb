```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
%matplotlib inline
```

# 画像の読み込みと色の話

```python
# 画像を読み込み
image = cv2.imread('./img/screen.png')

# 画像を表示する(BGR)
plt.imshow(image)
```

```python
# 変換する
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
```

```python
# gray scale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# グレースケールのとき、画像の表示はcmap='gray'でやらないとへんなことになる
plt.imshow(image_gray, cmap='gray')
```

```python
# 他にも色々ある

# HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(20, 10))
plt.subplot(1, 4, 1)
plt.imshow(image_hsv)

plt.subplot(1, 4, 2)
plt.imshow(image_hsv[:, :, 0]) # 色相

plt.subplot(1, 4, 3)
plt.imshow(image_hsv[:, :, 1]) # 彩度

plt.subplot(1, 4, 4)
plt.imshow(image_hsv[:, :, 2]) # 明度
```

```python
# 画像の中はこんな感じの行列です 
image.shape
```

1318行, 1760列, 3つの値（BGR）の値が入っています。

(x, y, 3)

```python
# 1000以下の行はすべて(255, 255, 255)を入れてあげると...

image_copy = image.copy()
image_copy[1000:, :, :] = (255, 255, 255)
plt.imshow(image_copy)
```

となる

# OpenCVの描画機能

http://labs.eecs.tottori-u.ac.jp/sd/Member/oyamada/OpenCV/html/py_tutorials/py_gui/py_drawing_functions/py_drawing_functions.html#opencv

```python
# 直線の描画

zeros = np.zeros_like(image, np.uint8)
# 画像, 線を引くpoint(x, y), 線が引き終わるポイント（x, y）, 色, 先の太さ
zeros2 = cv2.line(zeros, (0, 0), (511, 511), (255, 0, 0), 5)
plt.imshow(zeros2)
```

```python
# 画像, 線を引くpoint(x, y), 線が引き終わるポイント（x, y）, 色, 先の太さ
zeros2 = cv2.line(zeros, (0, 0), (1000, 1000), (255, 0, 0), 20)
plt.imshow(zeros2)
```

```python
# 長方形
# 画像, 四角の左上のポイント, 四角の右下のポイント, 色, pixel
zeros2 = cv2.rectangle(zeros, (384,0),(510,128),(0,255,0),3)
plt.imshow(zeros2)
```

```python
zeros2 = cv2.rectangle(zeros, (500,200),(1000,1000),(0,255,0),3)
plt.imshow(zeros2)
```

```python
# 円
# 画像, スタートするポイント, 半径?, 色, pixcelサイズ(-1だったら塗りつぶし)
zeros2 = cv2.circle(zeros,(447,63), 63, (0,0,255), -1)
plt.imshow(zeros2)
```

```python
# 円
# 画像、スタートのポイント, 半径？
zeros2 = cv2.circle(zeros,(600, 600), 200, (0,0,255), -1)
plt.imshow(zeros2)
```

400 ... 800までありそうなので、やっぱり半径っぽい
